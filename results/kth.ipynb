```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
from plot import *
from plot_utils import *
```

```python
dataset = 'kth'
prefix = '%s_'%(dataset+os.sep+dataset)
logs = [dataset+os.sep+f for f in os.listdir(dataset) if os.path.isfile(dataset+os.sep+f) and not f.endswith(('.json', '.py', '.ipynb'))]
logs.sort()
logs
```

```python
prefix2 = prefix+'2_'
f = plt.figure(figsize=(10.5,7.8))
ax = ipypthon_plot_log(logs, prefix=prefix2, suffix=('_val.log', '_train.log'), col=('acc_mean', 'acc'), figure=f,
                  labels=['Bayesian ResNet-18 Training', 'Bayesian ResNet-18 Validation', 'Frequentist ResNet-18 Training', 'Frequentist ResNet-18 Validation'],
                  linestyles=[':', '-', ':', '-'],
                  colors=[blue, blue, orange, orange],
                  legend_formatting=True,
                  xlim=(-5, 200), ylim=(0, 100), yticks=range(0,101,10))
ax.set_ylabel('Accuracy', fontsize=14, alpha=0.75)
ax.set_xlabel('Epochs', labelpad=15, fontsize=14, alpha=0.75)
plt.show()
```

### Split 2

```python
prefix2 = prefix + '2_'
f = plt.figure(figsize=(20,5))
ipypthon_plot_log(logs, prefix=prefix2, suffix='_val.log', col='acc_mean', subplot=131, figure=f, xlim=(-10, 500), ylim=(-2, 100), yticks=range(0,101,10))
ipypthon_plot_log(logs, prefix=prefix2, suffix='_train.log', col='acc', subplot=132, figure=f)
ipypthon_plot_log(logs, prefix=prefix2, suffix='_train.log', col='lr', subplot=133, figure=f)
plt.show()
```
