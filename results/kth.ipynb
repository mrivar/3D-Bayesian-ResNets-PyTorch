```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
from plot import *
from plot_utils import *
```

```python
dataset = 'kth'
prefix = '%s_'%(dataset+os.sep+dataset)
logs = [dataset+os.sep+f for f in os.listdir(dataset) if os.path.isfile(dataset+os.sep+f) and not f.endswith(('.json', '.py', '.ipynb'))]
logs.sort()
logs
```

```python
f = plt.figure(figsize=(21,7.8))
prefix2 = prefix+'2_'
ax = ipypthon_plot_log(logs, prefix=prefix2, suffix=('_val.log', '_train.log'), col=('acc_mean', 'acc'), subplot=121, figure=f,
                  #labels=['Bayesian ResNet-18 Training', 'Bayesian ResNet-18 Validation', 
                  #        'Dropout ResNet-18 Training', 'Dropout ResNet-18 Validation',
                  #        'Frequentist ResNet-18 Training', 'Frequentist ResNet-18 Validation'],
                  linestyles=[':', '-', ':', '-', ':', '-'],
                  colors=[blue, blue, green, green, orange, orange],
                  xlim=(-5, 200), ylim=(0, 100), yticks=range(0,101,10))
ax.set_ylabel('Accuracy', fontsize=14, alpha=0.75)
ax.set_xlabel('Epochs', labelpad=15, fontsize=14, alpha=0.75)

prefix4 = prefix+'4_'
#ax = ipypthon_plot_log(logs, prefix=prefix4, suffix=('5_val.log', '5_train.log', '8_val.log', '8_train.log'), col=('acc_mean', 'acc'), subplot=122, figure=f,
#ax = ipypthon_plot_log(logs, prefix=prefix4, suffix=('oth_val.log', 'oth_train.log'), col=('acc_mean', 'acc'), subplot=122, figure=f,
ax = ipypthon_plot_log(logs, prefix=prefix4, suffix=('8_oth_val.log', '8_oth_train.log', '5_val.log', '5_train.log'), col=('acc_mean', 'acc'), subplot=122, figure=f,
                  #labels=['Bayesian ResNet-18 Training', 'Bayesian ResNet-18 Validation', 
                  #        'Dropout ResNet-18 Training', 'Dropout ResNet-18 Validation',
                  #        'Frequentist ResNet-18 Training', 'Frequentist ResNet-18 Validation'],
                  linestyles=[':', '-', ':', '-', ':', '-'],
                  colors=[blue, blue, green, green, orange, orange],
                  xlim=(-5, 250), ylim=(0, 100), yticks=range(0,101,10))
ax.set_ylabel('Accuracy', fontsize=14, alpha=0.75)
ax.set_xlabel('Epochs', labelpad=15, fontsize=14, alpha=0.75)
plt.show()
```
