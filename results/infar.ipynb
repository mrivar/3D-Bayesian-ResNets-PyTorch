```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
from plot import *
from plot_utils import *
```

```python
dataset = 'infar'
prefix = '%s_'%(dataset+os.sep+dataset)
logs = [dataset+os.sep+f for f in os.listdir(dataset) if os.path.isfile(dataset+os.sep+f) and not f.endswith(('.json', '.py', '.ipynb'))]
logs.sort()
logs
```

```python
f = plt.figure(figsize=(21,7.8))
ax = ipypthon_plot_log(logs, prefix=prefix, suffix=('r_resnet18_oth_val.log', 'q-5_val.log', 'r_resnet18_oth_train.log', 'q-5_train.log', 'r_dropout_resnet18_train.log', 'r_dropout_resnet18_val.log'),
                       col=('acc_mean', 'acc'), subplot=131, figure=f,
                  #labels=['Bayesian ResNet-18 Training', 'Bayesian ResNet-18 Validation', 
                  #        'Dropout ResNet-18 Training', 'Dropout ResNet-18 Validation',
                  #        'Frequentist ResNet-18 Training', 'Frequentist ResNet-18 Validation'],
                  linestyles=[':', '-', ':', '-', ':', '-'],
                  colors=[blue, blue, green, green, orange, orange],
                  legend_formatting=True,
                  xlim=(-15,200), ylim=(0,102), yticks=range(0,101,10))
ax.set_ylabel('Accuracy', fontsize=14, alpha=0.75)
ax.set_xlabel('Epochs', labelpad=15, fontsize=14, alpha=0.75)

prefix4=prefix+'4_'
ax = ipypthon_plot_log(logs, prefix=prefix4, suffix=('oth_val.log', 'oth_train.log', '18_val.log', '18_train.log'),
                       col=('acc_mean', 'acc'), subplot=132, figure=f,
                  labels=['Bayesian ResNet-18 Training', 'Bayesian ResNet-18 Validation', 
                          'Dropout ResNet-18 Training', 'Dropout ResNet-18 Validation',
                          'Frequentist ResNet-18 Training', 'Frequentist ResNet-18 Validation'],
                  linestyles=[':', '-', ':', '-', ':', '-'],
                  colors=[blue, blue, green, green, orange, orange],
                  legend_formatting=True,
                  xlim=(-15,204), ylim=(0,102), yticks=range(0,101,10))
ax.set_ylabel('Accuracy', fontsize=14, alpha=0.75)
ax.set_xlabel('Epochs', labelpad=15, fontsize=14, alpha=0.75)

prefix5=prefix+'5_'
ax = ipypthon_plot_log(logs, prefix=prefix5, suffix=('val.log', 'train.log'),
                       col=('acc_mean', 'acc'), subplot=133, figure=f,
                  #labels=['Bayesian ResNet-18 Training', 'Bayesian ResNet-18 Validation', 
                  #        'Dropout ResNet-18 Training', 'Dropout ResNet-18 Validation',
                  #        'Frequentist ResNet-18 Training', 'Frequentist ResNet-18 Validation'],
                  linestyles=[':', '-', ':', '-', ':', '-'],
                  colors=[blue, blue, green, green, orange, orange],
                  legend_formatting=True,
                  xlim=(-15,204), ylim=(0,102), yticks=range(0,101,10))
ax.set_ylabel('Accuracy', fontsize=14, alpha=0.75)
ax.set_xlabel('Epochs', labelpad=15, fontsize=14, alpha=0.75)
plt.show()
```
