```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
from plot import *
from plot_utils import *
```

```python
infar = ['infar'+os.sep+f for f in os.listdir('infar') if os.path.isfile('infar'+os.sep+f) and not f.endswith(('.json', '.py', '.ipynb'))]
infar_prefix = '%s_'%('infar'+os.sep+'infar')
infar.sort()
ucfsports = ['ucfsports'+os.sep+f for f in os.listdir('ucfsports') if os.path.isfile('ucfsports'+os.sep+f) and not f.endswith(('.json', '.py', '.ipynb'))]
ucfsports_prefix = '%s_'%('ucfsports'+os.sep+'ucfsports')
ucfsports.sort()
ucf11 = ['ucf11'+os.sep+f for f in os.listdir('ucf11') if os.path.isfile('ucf11'+os.sep+f) and not f.endswith(('.json', '.py', '.ipynb'))]
ucf11_prefix = '%s_'%('ucf11'+os.sep+'ucf11')
ucf11.sort()
jhmdb = ['jhmdb'+os.sep+f for f in os.listdir('jhmdb') if os.path.isfile('jhmdb'+os.sep+f) and not f.endswith(('.json', '.py', '.ipynb'))]
jhmdb.sort()
kth = ['kth'+os.sep+f for f in os.listdir('kth') if os.path.isfile('kth'+os.sep+f) and not f.endswith(('.json', '.py', '.ipynb'))]
kth_prefix = '%s_'%('kth'+os.sep+'kth')
kth.sort()
```

```python
f = plt.figure(figsize=(11,9))

prefix2 = (kth_prefix+'2_BBB', kth_prefix+'2_resnet18')
ax = ipypthon_plot_log(kth, prefix=prefix2, suffix=('_val.log', '_train.log'), col=('acc_mean', 'acc'),figure=f,
                  labels=['Bayesian ResNet-18 Training', 'Bayesian ResNet-18 Validation', 
                          #'Dropout ResNet-18 Training', 'Dropout ResNet-18 Validation',
                          'Frequentist ResNet-18 Training', 'Frequentist ResNet-18 Validation'],
                  linestyles=[':', '-', ':', '-', ':', '-'],
                  colors=[blue, blue, orange, orange],#[blue, blue, green, green, orange, orange],
                  legend_fontsize=15,
                  xlim=(-5, 161), ylim=(0, 100), yticks=range(0,101,10))
ax.set_ylabel('Accuracy', fontsize=14, alpha=0.75)
ax.set_xlabel('Epochs', labelpad=15, fontsize=14, alpha=0.75)
ax.set_title('KTH', fontsize=18)

plt.show()
```
